```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
lena_gray = cv2.imread("./images/Lena_gray.jpg", cv2.IMREAD_GRAYSCALE)
# lena_gray = cv2.imread("./Lena_grey.jpg",0)
print("data type of the image: ", lena_gray.dtype)
print("shape of the image: ", lena_gray.shape)
print("value of lena_gray[0,0]: ", lena_gray[0, 0])
plt.imshow(lena_gray, cmap='gray')
```

```python
img = np.zeros(shape=(512, 512, 3), dtype=np.uint8) # 3?? each one stads for RGB channel?
print("shape of the img: ", img.shape)
plt.subplot(1,3,1)
plt.title("all zero")
plt.imshow(img)

for i in range(512):
    for j in range(512):
        img[i, j, 0] = 255
plt.subplot(1,3,2)
plt.title("red colour")
plt.imshow(img)

for i in range(100):
    for j in range(100):
        img[i,j,0] = 0
        img[i,j,1] = 128
        img[i,j,2] = 128
plt.subplot(1,3,3)
plt.title("other colour")
plt.imshow(img)
```

```python
flowers_RGB = cv2.imread("./images/Flowers.jpg", cv2.IMREAD_COLOR)
flowers_RGB = cv2.cvtColor(flowers_RGB, cv2.COLOR_BGR2RGB)
print("data type of the image: ", flowers_RGB.dtype)
print("shape of the image: ", flowers_RGB.shape)
print("value of the flowers_RGB[0,0]: ", flowers_RGB[0,0])
plt.imshow(flowers_RGB)
```

```python
plt.imshow(flowers_RGB)

flowers_R = flowers_RGB[:,:,0]
flowers_G = flowers_RGB[:,:,1]
flowers_B = flowers_RGB[:,:,2]

print("shape of the flowers_RGB: ", flowers_RGB.shape)
print("shape of the flowers_RGB(single channel): ", flowers_R.shape)

plt.figure()

plt.subplot(1,3,1)
plt.title("R channel")
plt.imshow(flowers_R, cmap='gray')

plt.subplot(1,3,2)
plt.title("G channel")
plt.imshow(flowers_G, cmap='gray')

plt.subplot(1,3,3)
plt.title("B channel")
plt.imshow(flowers_B, cmap='gray')
```

```python
lena_gray = cv2.imread("./images/Lena_gray.jpg", cv2.IMREAD_GRAYSCALE)

plt.subplot(1,3,1)
plt.title("original img")
plt.imshow(lena_gray, cmap='gray')

h, w = lena_gray.shape
lena_add = np.zeros((h,w), np.uint8)

for i in range(h):
    for j in range(w):
        lena_add[i,j] = lena_gray[i,j] + 50

plt.subplot(1,3,2)
plt.title("add 50")
plt.imshow(lena_add, cmap='gray')

lena_add = lena_gray + 100

for i in range(h):
    a = True
    for j in range(w):
        if(lena_add[i,j] > 255):
            print(lena_add[i,j])
            a = False;
    if not a: break;

plt.subplot(1,3,3)
plt.title("add 100")
plt.imshow(lena_add, cmap='gray')
```

```python
lena_gray = cv2.imread("./images/Lena_gray.jpg", cv2.IMREAD_GRAYSCALE)

plt.subplot(1,3,1)
plt.title("original")
plt.imshow(lena_gray, cmap='gray')

lena_mul = np.int32(lena_gray)*2
lena_mul[lena_mul>255] = 255
lena_mul = np.uint8(lena_mul)

plt.subplot(1,3,2)
plt.title("mul (saturated)")
plt.imshow(lena_mul, cmap='gray')

lena_sub = np.int32(lena_gray) - 100
lena_sub[lena_sub < 0] = 0
lena_sub = np.uint8(lena_sub)

plt.subplot(1,3,3)
plt.title("sub (saturated)")
plt.imshow(lena_sub, cmap='gray')
```

```python
lena = cv2.imread("./images/lena_color.jpg")
lena = cv2.cvtColor(lena, cv2.COLOR_BGR2RGB)

pattern = cv2.imread("./images/Pattern1.png")
pattern = cv2.cvtColor(pattern, cv2.COLOR_BGR2RGB)

alpha = 0.5
beta = 1 - alpha
mixed = (alpha*lena)+(beta*pattern)
mixed = np.uint8(mixed)

plt.subplot(1,3,1)
plt.imshow(lena)
plt.subplot(1,3,2)
plt.imshow(pattern)
plt.subplot(1,3,3)
plt.imshow(mixed)
```

```python
img1 = cv2.imread("./images/Pattern1.png", 0)
img2 = cv2.imread("./images/Pattern2.png", 0)

img_diff = np.abs(img1 - img2)

plt.subplot(1,2,1)
plt.imshow(img1, cmap='gray', vmin=0, vmax=255)
plt.subplot(1,2,2)
plt.imshow(img2, cmap='gray', vmin=0, vmax=255)

plt.figure()
plt.imshow(img_diff, cmap='gray', vmin=0, vmax=255)
plt.colorbar()
```
