```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def make_hist(image):
    assert len(image.shape) == 2, "grayscale? "
    histogram = np.zeros((256), dtype=np.int32)
    h, w = image.shape
    
    for i in range(h):
        for j in range(w):
            histogram[i, j] = image[i, j];
    return histogram
```

```python
lena = cv2.imread("./images/week3/lena.png", 0)
plt.imshow(lena, cmap='gray', vmin=0, vmax=255)

plt.figure()
lena_hist = make_hist(lena)

x = np.arange(256)
plt.bar(x, height=lena_hist)

plt.title("historam")
plt.xlabel("value")
plt.ylabel("frequency")
plt.show()
```
