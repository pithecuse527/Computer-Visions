```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("rhino.png")
h, w, c =img.shape

plt.figure()
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

markers = cv2.imread("markers.png", 0)
markers = np.int32(markers)

plt.figure()
plt.imshow(markers)
plt.colorbar()

result = cv2.watershed(img, markers)

plt.figure()
plt.imshow(result)
plt.colorbar()
```

```python
markers = np.zeros((h,w), np.int32)

radius = 10
thickness = -1

markers_bg =[(10,10), (600,100), (1000,200)]
color_bg =1
for point in markers_bg:
    cv2.circle(markers, point , radius, color_bg, thickness)
    
markers_rhino = [(400,500), (450,400), (500,300), (600,450), (800,300)]
color_rhino = 2 
for point in markers_rhino:
    cv2.circle(markers, point , radius, color_rhino, thickness)

markers_tree = [(100,300), (400,250), (700,200), (1050,400)]
color_tree = 3
for point in markers_tree:
    cv2.circle(markers, point , radius, color_tree, thickness)

markers_grass = [(100,600), (400,620), (700,550), (1000,520), (1100, 520)]
color_grass = 4
for point in markers_grass:
    cv2.circle(markers, point , radius, color_grass, thickness)

plt.imshow(markers)
plt.colorbar()
```

```python
img = cv2.imread("rhino.png")
h, w, c =img.shape

num_iterations = 4
prior = 2
num_levels = 4
num_histogram_bins = 5
num_superpixels = int(h*w / 2000)

seeds = cv2.ximgproc.createSuperpixelSEEDS(w, h, c, num_superpixels, num_levels, prior, num_histogram_bins)

seeds.iterate(img, num_iterations)

num_of_superpixels_result = seeds.getNumberOfSuperpixels()
print('Start number of superpixels: %d' % num_superpixels)
print('Final number of superpixels: %d' % num_of_superpixels_result)

labels = seeds.getLabels()

color_img = np.zeros((h, w, 3), np.uint8)
color_img[:] = (0, 0, 255)

mask = seeds.getLabelContourMask(False)

mask_inv = cv2.bitwise_not(mask)
result_bg = cv2.bitwise_and(img, img, mask=mask_inv)
result_fg = cv2.bitwise_and(color_img, color_img, mask=mask)
result = cv2.add(result_bg, result_fg)

result = cv2.cvtColor(result, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(20,20))
plt.imshow(result)
```
